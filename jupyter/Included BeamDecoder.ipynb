# Same thing with odtm modification

```python
import pickle
import codecs

with open(join(data_dir, "most-common-editions-odtm-0.3.pickle"), "rb") as f:
    sorted_editions = pickle.load(f)
    
def sources_words_from_file(path):
    source_words = []

    with codecs.open(path, 'r', encoding='utf-8') as f:
        for line in f:
            src = line.strip().lower()
            src_words = src.split(' ')
            assert len(src_words) > 0
            source_words.append(src_words)
    return source_words

source_words = sources_words_from_file(join(data_dir, file_name))

examples = []
for source_word_list in source_words:
    for sorted_edition in sorted_editions:
        ex = EditExample(source_words=source_word_list, insert_words=[], insert_exact_words=sorted_edition[1][u"insert"], delete_words=[], delete_exact_words=sorted_edition[1][u"delete"], target_words=[], edit_embed=None)
        examples.append(ex)
        
beam_list, edit_traces = exp.editor.edit(examples, max_seq_length=35, beam_size=5, batch_size=256)

l = []
candidates = []
for edit_trace in edit_traces:
    for candidate in edit_trace.decoder_trace.candidates:
        if u'<unk>' not in candidate.sequence:
            if edit_trace.example.source_words != candidate.sequence:
                if candidate.sequence not in candidates:
                    l.append({"source": edit_trace.example.source_words, "generation": candidate.sequence, "prob": candidate.prob})
                    candidates.append(candidate.sequence)
                    
sorted_l = sorted(l, key=lambda x: x["prob"], reverse=True)

with open(join(data_dir, "SCinternal_generated_0.0.2"), "wb") as f:
    for x in sorted_l:
        line = "(%.2f) - %s \n" % (x["prob"], " ".join(x["generation"]))
        f.write(line.encode("utf8"))
```

# Reverse Same thing with odtm modification

```python
# Constant definitions

# Folder Structures
DATA_DIR = join(data.workspace.root, "textgen_dataset")
OTDM_DATA_DIR = join(DATA_DIR, "scinternal", "otdm")

# Files
END_FILE_NAME = "_preprocessed-0.5.tsv" # change here
GENERATED_FILE = "most-common-editions-odtm-0.5-preprocessed.pickle" # change here

FILES_NAME = ["valid"+END_FILE_NAME, "test"+END_FILE_NAME, "train"+END_FILE_NAME]
```

```python
# word vocab
word_vocab = exp.editor.train_decoder.word_vocab
```

```python
examples = []

for FILE_NAME in FILES_NAME:
    path = join(OTDM_DATA_DIR, FILE_NAME)
    with open(path, "rb") as f:
        for line in f:
            target, source = line.strip().split("\t")
            target_words = target.split(" ")
            source_words = source.split(" ")

            insert_words = sorted(set(target_words) - set(source_words))
            delete_words = sorted(set(source_words) - set(target_words))

            insert_words = [insert_word for insert_word in insert_words if word_vocab.word2index(insert_word) != 0]
            delete_words = [delete_word for delete_word in delete_words if word_vocab.word2index(delete_word) != 0]

            examples.append(EditExample(source_words=source_words, insert_words=[], insert_exact_words=insert_words,
                               delete_words=[], delete_exact_words=delete_words, target_words=target_words))
```

```python
len(examples)
```

```python
beam_list, edit_traces = exp.editor.edit(examples, max_seq_length=35, beam_size=5, batch_size=256)

l = []
candidates = []
for edit_trace in edit_traces:
    for candidate in edit_trace.decoder_trace.candidates:
        if u'<unk>' not in candidate.sequence:
            if edit_trace.example.source_words != candidate.sequence:
                if candidate.sequence not in candidates:
                    l.append({"source": edit_trace.example.source_words, "generation": candidate.sequence, "prob": candidate.prob})
                    candidates.append(candidate.sequence)
                    
sorted_l = sorted(l, key=lambda x: x["prob"], reverse=True)


with open(join(DATA_DIR, "SCinternal_generated_0.0.7"), "wb") as f:
    f.write("trg = small dataset preprocessed, source = big dataset, remove unknown insert/delete words, jaccard 0.5\n\n")
    for x in sorted_l:
        line = "(%.2f) - %s \n" % (x["prob"], " ".join(x["generation"]))
        f.write(line.encode("utf8"))
```
